```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
credits = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv") 
```

```python
credits = credits.drop("title", axis=1)
# movies["title"] = movies["title"].apply(lambda x: x.lower())
```

```python
movies = pd.concat([credits, movies], axis=1)
```

```python
# dropping columns

# Imp columns

imp_cols = [
    "id",
    "title",
    "cast",
    "crew",
    "genres",
    "keywords",
    "overview",
#     "release_date",
]

movies = pd.DataFrame(movies)

movies = movies[imp_cols]
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna(axis=0)
```

```python
movies.isna().any()
```

```python
movies.duplicated().any()
```

```python
movies.genres[0]
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): # to change a list of string into a list ast.literal_eval() function is needed
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies.genres.apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.cast[0]
```

```python
# we will select the top 5 cast of that movie

def convert5(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=5:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert5)
```

```python
# pd.set_option("display.max_colwidth", None)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
movies.rename(columns= {"crew" : "director"}, inplace=True)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["director"] = movies["director"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["director"]
```

```python
movies.head()
```

```python
df = movies[["id","title", "tags"]]
```

```python
df.head()
```

```python
df["tags"] = df["tags"].apply(lambda x: " ".join(x))
```

```python
df.head()
```

```python
df["tags"] = df["tags"].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
# we need to apply stemming techniques to extract the original word from a list of words

#for example
#  {loved, loving, love} to {love, love, love}

# !pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df["tags"] = df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3500, stop_words="english")
```

```python
vectors = cv.fit_transform(df["tags"]).toarray()
```

```python
# now we have to calculate the cosine distance

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
#     movie  = movie.lower()
    movie_index = df[df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
